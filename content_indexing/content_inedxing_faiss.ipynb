```python
import pandas as pd
import numpy as np
```

```python
data_folder = '/Users/cep4u/JingEdward/tunen/data/content_rawdata'
filename = 'spotify_audio_content_db_2M.csv'
file_path = f"{data_folder}/{filename}"

df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.iloc[:,9:20]
```

```python
one_hot = pd.get_dummies(df['key'], prefix='key').astype(int)

df = df.drop('key', axis=1)
df = pd.concat([df, one_hot], axis=1)
```

```python
one_hot = pd.get_dummies(df['mode'], prefix='mode').astype(int)

df = df.drop('mode', axis=1)
df = pd.concat([df, one_hot], axis=1)
```

```python
df['loudness_scaled'] = (df['loudness'] - df['loudness'].mean()) / df['loudness'].std()
```

```python
df['scaled_tempo'] = (df['tempo'] - df['tempo'].min()) / (df['tempo'].max() - df['tempo'].min())
```

```python
df.columns
```

```python
columns_ = ['danceability','energy', 
            'speechiness', 'acousticness', 'instrumentalness',
            'liveness', 'valence','key_0', 'key_1',
           'key_2', 'key_3', 'key_4', 'key_5', 'key_6', 'key_7', 'key_8', 'key_9',
           'key_10', 'key_11', 'mode_0', 'mode_1', 'loudness_scaled',
           'scaled_tempo'
           ]
```

```python
data = df[columns_]
```

```python
from sklearn.neighbors import NearestNeighbors

# Assuming 'data' is your preprocessed dataset
knn_model = NearestNeighbors(algorithm='kd_tree', n_neighbors=100)
knn_model.fit(data)
```
