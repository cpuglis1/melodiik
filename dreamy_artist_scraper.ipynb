```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import urllib.parse
```

```python
def get_artist_names(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    artist_names = []
    
    for a in soup.find_all('a', href=True):
        if a['href'].startswith('/music/'):
            artist_name = a['href'].replace('/music/', '')
            artist_names.append(artist_name)

    unique_names = clean_names(artist_names)
    
    return unique_names
```

```python
def clean_names(artist_names):
    
    cleaned_names = [urllib.parse.unquote(name).replace('+', ' ') for name in artist_names if name != '+free-music-downloads']
    
    unique_names = list(dict.fromkeys(cleaned_names))
    
    return unique_names
```

## Bedroom Pop

```python
base_link = 'https://www.last.fm/tag/bedroom+pop/artists?page='

all_bedroom_pop = []

for i in range(87):
    url = base_link + str(i+1)
    all_bedroom_pop.extend(get_artist_names(url))
    
```

## Indie

```python
base_link = 'https://www.last.fm/tag/indie+rock/artists?page='

all_indie = []

for i in range(477):
    url = base_link + str(i+1)
    all_indie.extend(get_artist_names(url))
```

```python
## Combine in single DF
```

```python
data = [
    (artist, 'indie') for artist in all_indie
] + [
    (artist, 'indie rock') for artist in all_indie_rock
] + [
    (artist, 'dream pop') for artist in all_dream_pop
] + [
    (artist, 'bedroom pop') for artist in all_bedroom_pop
]

df = pd.DataFrame(data, columns=['artist', 'genre'])

df = df.groupby('artist')['genre'].apply(', '.join).reset_index()

print(df)
```
